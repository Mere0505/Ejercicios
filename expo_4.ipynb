Medir distancias en el espacio, especialmente a escalas astronómicas como estrellas y galaxias distantes, presenta desafíos únicos. Aquí hay algunas de las técnicas utilizadas para medir distancias en el universo:

**Paralaje Estelar:**
Esta es una técnica común para medir distancias a estrellas relativamente cercanas. Para galaxias distantes, el corrimiento al rojo del espectro de luz (el llamado efecto Doppler) se utiliza para medir la velocidad a la que se alejan de nosotros. **La tensión de Hubble**

La "tensión de Hubble" subraya una incertidumbre fundamental: Aunque es un hecho que el cosmos se expande en todas direcciones, los científicos aún luchan por determinar la velocidad exacta de esta expansión. El hecho de que aún no podamos calcularlo con precisión sugiere que puede haber una variable aún desconocida en nuestra comprensión. Es un recordatorio de que todavía estamos en las primeras etapas de entender completamente el universo.
Aunque a nosotros nos parezca inalterable, todas las cosas del universo en realidad se están alejando de nosotros a una velocidad calculada como la "Constante de Hubble". Esta velocida, en la que concuerdan las mediciones de los telescopios Webb y Hubble, sería de unos 73 kilómetros por segundo por megaparsec.

La Constante de Hubble es una medida de la velocidad a la que se están alejando o acercando entre sí las galaxias en el universo. Recibe su nombre en honor al astrónomo Edwin Hubble, quien realizó observaciones fundamentales que llevaron al descubrimiento de la expansión del universo.
La expansión del universo se describe mediante la Ley de Hubble, que establece que la velocidad de recesión de una galaxia es proporcional a su distancia. La ecuación básica es:

$v=H_0 \cdot d$

Donde:

$v$ es la velocidad de recisión de la galaxia.

$H_0$ es la Constante de Hubble, que representa la tasa de expansión del universo.

$d$ es la distancia a la galaxia.

En términos más sencillos, la Constante de Hubble indica cuánto se expande el universo en un determinado momento. Su valor se expresa típicamente en unidades de velocidad divididas por distancia, como kilómetros por segundo por megapársec (km/s/Mpc).

Un teodolito es un instrumento de medición utilizado en topografía y geodesia para medir ángulos verticales y horizontales con gran precisión. Este dispositivo es esencial en trabajos de levantamiento topográfico, construcción y cartografía.

**Funcionamiento del Teodolito:**

**Mira y Luneta:**
El teodolito consta de una mira, que es una línea de referencia vertical, y una luneta, que es un sistema óptico que puede girar en dos ejes perpendiculares entre sí. La luneta tiene un retículo que ayuda a apuntar con precisión a puntos específicos.

**Medición de Ángulos:**
Los teodolitos modernos suelen tener lecturas digitales, lo que facilita la lectura precisa de los ángulos. Algunos modelos también tienen compensadores electrónicos para corregir pequeñas irregularidades causadas por movimientos o vibraciones.

**Márgenes de Error:**

**Error de Lectura:**

La precisión también depende de la habilidad y la experiencia del operador en el manejo del instrumento.

**Ruido Blanco:**

El ruido blanco es una señal aleatoria que tiene una densidad espectral de potencia constante a lo largo de todas las frecuencias. Esto significa que la energía del ruido es uniformemente distribuida en el espectro de frecuencias. En términos más simples, todas las frecuencias tienen la misma cantidad de energía. El término "blanco" proviene de la analogía con la luz blanca, que contiene todas las longitudes de onda visibles en el espectro electromagnético.

En el dominio del tiempo, el ruido blanco se manifiesta como una secuencia de valores aleatorios independientes e igualmente distribuidos. Es importante destacar que el ruido blanco es un concepto idealizado, ya que en la práctica no hay señales que sean completamente "blancas" en todas las frecuencias.

**Ruido Gaussiano:**

El ruido gaussiano, también conocido como ruido normal o ruido gaussiano blanco, es un tipo específico de ruido que sigue una distribución normal (o gaussiana). En este tipo de ruido, las amplitudes de las señales en el dominio del tiempo siguen una distribución de campana de Gauss. Esto significa que las fluctuaciones en el ruido son más probables alrededor de un valor medio, y las amplitudes más extremas son menos probables.

El ruido gaussiano es particularmente relevante porque surge naturalmente en una variedad de procesos físicos y se comporta de manera predecible estadísticamente. Muchos fenómenos naturales, como las pequeñas variaciones en la temperatura, la presión atmosférica y otros factores, pueden modelarse como ruido gaussiano.

Despreciar los términos de orden superior en la expansión en serie de la raíz cuadrada ya que $\frac{e^2_k}{L}$ es pequeño. Dado que $l_k$ es $L$, la longitud del tramo $kth$ se **sobreestima** en una cantidad $\frac{e^2_k}{L}$. Para $n$ longitudes de cinta, el error total es:

Sin embargo, parecería difícil establecer un modelo realista que tenga esto en cuenta; por lo tanto, los errores aleatorios constituyen una muestra aleatoria de una distribución normal con media cero y varianza $\sigma^2$, que se supone conocida.
Para obtener el comportamiento estadístico de $\delta$ necesitamos conocer el comportamiento del $e^2_i$. Supongamos que tenemos "m" números $x_i$ con media cero y varianza $\sigma^2$,

$\sigma^2=\frac{1}{m} \sum_{i=1}^m x^2_i$

Si denotamos la media de $x^2_i$ por $X$, entonces:

$X=\frac{1}{m} \sum_{i=1}^m x^2_i= \sigma^2$

y la varianza de $x^2_i$ viene dada por:
$v^2=\frac{1}{m} \sum_{i=1}^m (x^2_i-X)^2=\frac{1}{m} \sum_{i=1}^m x^4_i - \frac{2X}{m} \sum_{i=1}^m x^2_i +X^2= \frac{1}{m}\sum_{i=1}^m x^4_i- \sigma^4$

La suma de $x^4_i$ dividida por $m$ en la última línea de la ecuación es simplemente el cuarto momento de $x_i$. Si ahora se supone que $x_i$ normalmente se distribuyen con media cero y varianza $\sigma^2$, el cuarto momento es $3\sigma^4$ (ver más abajo), por lo que la varianza de $x^2_i$ viene dada por:
$v^2=3 \sigma^4-\sigma^4=2 \sigma^4$

Un método más sofisticado para obtener este resultado es el siguiente. Supongamos que la distribución de frecuencia de una variable $x$ viene dada por $f(x)$, $- \infty < x < \infty$, de modo que su función de distribución acumulativa es

$F(\xi)=\int_{-\infty}^\xi f(x)dx$

Cuando $F(\xi)$ es la probabilidad de que $x$ sea menor que $\xi$, ahora consideramos la distribución de frecuencia de $y=x^2$. La probabilidad de que $y$ sea menor que, digamos $\eta (\eta >0)$ es la probabilidad de que
$-\sqrt{\eta}< x <\sqrt{\eta}$ esto es:

$G(\eta)=\int_{-\sqrt{\eta}}^{+\sqrt{\eta}} f(x)dx$

Donde $G(\eta)$ es la función de distribución acumulativa de $y$. La función de distribución de frecuencia para $y$ es:

$\frac{dG}{d \eta}_{n=y}=\frac{f(\sqrt{y})+f(-\sqrt{y})}{2\sqrt{2}}=g(y)$

Donde esto define $g(y)$. La media $\overline{y}$ de $y$ está dada por:

$\overline{y}=\int_{0}^{\infty}yg(y)dy=\frac{1}{2}\int_{0}^{\infty}y^{\frac{1}{2}}\left[ \frac{f(\sqrt{y})+f(-\sqrt{y})}{2\sqrt{2}}\right] dy$

$=\int_{0}^{\infty}u^{2}\left[ f(u)+f(-u)\right]du= \int_{-\infty}^{\infty}u^{2}f(u)du=\mu_2$

Donde $\mu_2$ denota el segundo momento de $f(x)$ con respecto al origen. La varianza $v^2$ de $y$ está dada por:

$\overline{y}=\int_{0}^{\infty}(y-\overline{y})^2 g(y)dy$

$=\frac{1}{2}\int_{0}^{\infty} (y^{\frac{3}{2}}-2y^\frac{1}{2}\overline{y}+\overline{y}^2y^{-\frac{1}{2}})\left[ f(\sqrt{y})+f(-\sqrt{y})\right]dy$

$=\int_{-\infty}^{\infty} u^4f(u)du-2\mu_2\overline{y}+\overline{y}^2=\mu_4 - \mu^2_2$

Donde $\mu_4$ denota el cuarto momento de $f(x)$ con respecto al origen. Para la distribución normal tenemos:

$f(x)=\frac{1}{(2\pi)^{\frac{1}{2}}\sigma} e^{\frac{1}{\sigma}^2/2}$

Y es fácil demostrar que $\mu_2= \sigma^2$ y $\mu_4 = 3 \sigma^4 $, lo que conduce a $y= \sigma ^2$ y $v^2= 2 \sigma ^4$, como antes.
Volvemos ahora al problema original. Entonces la media de $ e_r $ es cero. Denota la varianza de $e_r$ por $\sigma^2$. Luego, a partir de los resultados de los dos últimos párrafos, la media de $e^2_r$ es $\sigma^2$ y su varianza es $2 \sigma^4$. La cantidad $\delta$ en (13.14) es la suma de $n$ variables independientes. Según el teorema del límite central, $\delta$ se comportará, para $n$ grandes, como una variable normal con:

Media $=\frac{1}{2L}n$ (media de $e^2_r$) $=\frac{n\sigma^2}{2L}$

Varianza $=\frac{1}{4L^2}n$ (varianza de $e^2_i$) $=\frac{2n\sigma^4}{4L^2}$

Desvación estándar $(desviación estandar)^\frac{1}{2}=\frac{(2n)^{\frac{1}{2}}\sigma^2}{2L}$

Estas fórmulas constituyen la respuesta requerida.

En el caso específico mencionado en el primer párrafo de la sección tenemos $L=100$, $n=\frac{8000}{100}=80$, y la desviación estándar de $e_r$ es $0.5$; es decir, $\sigma^2=(0,5)^2=0,25$. Por eso

Media$=\frac{(80)(0,25))}{200}=0,1$,

Desviación estándar $=\frac{(160)^{\frac{1}{2}}(0,25)}{200}=0,0158$

Los límites de confianza de $95 \%$ son $1,95$ multiplicado por la desviación estándar, o $(1,95)(0,0158)=0,03 ft$. Por lo tanto, el error se encuentra en el rango $0.1 \pm 0.03$ ft con $95 \%$ de probabilidad, como se citó anteriormente.

**una pregunta importante de esté ejercicio**


**¿Por qué se distribuyen los errores de manera normal?**

La suposición de que los errores se distribuyen de manera normal es una simplificación común en muchos modelos estadísticos y análisis debido al Teorema del Límite Central. Este teorema establece que, para una muestra lo suficientemente grande, la distribución de la media muestral de una variable aleatoria, independientemente de su distribución subyacente, se aproximará a una distribución normal.

Se asume que los errores de alineación $e_1, e_2,..., e_n$ son pequeños y aleatorios. Cuando se suman muchos pequeños efectos aleatorios, la distribución resultante tiende a seguir una distribución normal debido al Teorema del Límite Central. Esta suposición facilita el análisis estadístico y proporciona herramientas bien establecidas, como la distribución normal, para modelar y entender la variabilidad en los datos.


**Descripción del problema**

Se tiene una cinta de acerro de longitud $L$ utilizada para medir distancias entre dos puntos. Debido a errores de alineación, la distancia observada $l_k$ es mayor que la distancia real. Se asume que los errores $e_k$ son pequeños, aleatorios y tienen una desviación estnadar de $0.5ft$. Se desea demostrar que el error total es una distancia de $8,000ft$ entre $0.07 ft$ y $0.13 ft$ con una probabilidad de $95 \%$.

**Método de medición**
Se utiliza fórmula $l_k \approx L-\frac{e_k}{2L}$ para representar la longitud medida $l_k$ en función de la longitud real $L$ y el error de alineació  $e_k$.  Se simplifica la expresión despreciando términos de orden superior en la expresión en serie de la raíz cuadrada, ya que $\frac{e_k}{2L}$ es pequeño.

**Error total**
El error total $\delta$ para $n$ mediciones se define como $\delta= \frac{1}{2L} \sum_{r=1}^n e^2_r$ y asume que los errores aleatorios $e_1, ... , e_n$ son independientes.

**Análisis estadístico**
Se asume que los errores $e_r$ siguen una distribución normal con media cero y varianza conocida $\sigma^2 $. Se utiliza el teorema del límite central para obtener el comportamiento estadístico de $\delta$. La media y la varianza de $\delta$ se derivan en función de $\sigma^2$ y $L$.

**Resultados**
Para $n$ grande, $\delta$ se comporta como una variable normal con media $\frac{n\sigma^2}{2L}$ y varianza $\frac{2n\sigma^4}{4^2L}$, la desviación estándar de $\delta$ es $\frac{\sqrt{2n}\sigma^2}{2L}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del problema
L = 100  # Longitud de la cinta en pies
n = 80   # Número de mediciones
sigma = 0.5  # Desviación estándar de los errores de alineación

# Simulación de errores aleatorios
errores = np.random.normal(0, sigma, n)

# Cálculo de la cantidad delta según la ecuación 13.14
delta = (1/(2*L)) * np.sum(errores**2)

# Impresión de resultados
print(f'Media de errores al cuadrado: {np.mean(errores**2)}')
print(f'Varianza de errores al cuadrado: {np.var(errores**2)}')
print(f'Cantidad delta: {delta}')

# Visualización de la distribución de errores
plt.hist(errores, bins=20, density=True, alpha=0.7, color='blue', label='Errores')
plt.xlabel('Error de alineación (pies)')
plt.ylabel('Frecuencia')
plt.title('Distribución de errores de alineación')
plt.legend()
plt.show()
```
