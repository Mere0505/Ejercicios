Un fabricante de shampú dispone de $\$15$ millones de pesos para producir y promover un nuevo producto y promover un nuevo producto que evita la caida del cabello. Estima que si gasta $x$ millones de pesos en producirlo, $y$  millones de pesos para anunciarlo, entonces podrá vender aproximadamente: $z=12y\sqrt{x^3}$ miles de frascos de su producto en el curso del año.
¿Cómo debe distribuir su prosupuesto en precio y publicidad si quiere maximizar en sus ventas?

**Solución:**

**programa de opt.**

*Variable de decisión:*

$x=$"Costo de producción"

$y=$"Costo de anuncio"

*Restricciones*

$x+y+15, x,y \geq0$

*Función objetivo*

$z=12y\sqrt{x^3}$ (Maximizar)

```python
%matplotlib inline
```

```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
import pylab as pl
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = pl.figure(figsize=(10,8))
ax = Axes3D(fig)
X = np.arange(0, 10, 0.25)
Y = np.arange(0, 10, 0.25)
X, Y = np.meshgrid(X, Y)

Z = 12*Y*X**(3/2)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap="inferno")
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=pl.cm.hot)
ax.set_zlim(0,2500)

pl.show()
```

```python
def f(x):
    return 12.0*x[1]*(x[0])**(3.0/2.0)
```

```python
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y)
Z = f(np.vstack([X.ravel(), Y.ravel()])).reshape((100,100))
plt.figure(figsize=(10,8))
plt.contour(X, Y, Z, np.arange(0,2500, 50), cmap='jet');
plt.plot(x, 15.0 - x, 'k:', linewidth=3)
plt.xlabel('$x$'); plt.ylabel('$y$');
plt.title("Proyección en el plano XY")
plt.axis([0.0,10.0,0.0,10.0])
```

```python
def f(x):
    return -12.0*x[1]*(x[0])**(3.0/2.0)
```

```python
cons = ({'type': 'eq',
         'fun' : lambda x: np.array([x[0] + x[1]-15.0])})#,
         #'jac' : lambda x: np.array([1.0, 2.0*x[1]])})

bnds = ((0, 10.0), (0.0, 10.0))
```

```python
x0 = [1.0, 4.0]
```

```python
cx = opt.minimize(f, x0, bounds=bnds, constraints=cons)
cx
```
