**a)** Max $x_{2}+y_{2}+y-1$ sujeta a $x_{2}+y_{2} \leq 1$

Primero realice un bosquejo para resolverlo.

Para maximizar la función $x_{2}+y_{2}+y-1$ sujeta a $x_{2}+y_{2} \leq 1$, notamos que la función objetivo es una parábola que abre hacia arriba y la restricción es un círculo con centro en el origen, con radio **1**.

La función objetivo aumenta a medida que nos alejemos del origen en dirección positiva en **y**.
Entonces el punto más alejado del origen en la circunferencia será aquel en que **y** sea máximo y $x=0$.

El punto en la circunferencia que maximiza $y$ es $(0,1)$ entonces el valor máximo de la función es:

$f(0,1)=0^{2}+1^{2}+1-1=1$


Ahora realizamos el código en python.

```python
%matplotlib inline
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
```

Ahora vamos hacer un programa donde calculé nuestro resultado.

```python
import numpy as np
import scipy.optimize as opt

# Definimos la función a maximizar
def f(x):
    return (x[0]**2 + x[1]**2 + x[1] - 1)

# Restricciones de desigualdad
def constraint1(x):
    return x[0] + x[1]**2 - 1.0

# Condiciones iniciales
x0 = [1, 1]

# Restricciones y límites de las variables
cons = ({'type': 'ineq', 'fun': constraint1})
bnds = ((-10, 10), (-10, 10))

# Resolvemos la optimización con restricciones
result = opt.minimize(f, x0, bounds=bnds, constraints=cons)

result
```

**b)** Max $xyz$ sujeta a $x^{2}+y^{2}+z^{2} \leq 1$

Primero realice un bosquejo, entonces l máximo del producto*xyz* sujeto a $x^{2}+y^{2}+z^{2} \leq 1$ de que el punto$(x,y,z)$, está dentro de la esfera unitaria centrada en el origen. Dado que $xyz$ es positivo, usaremos el valor absoluto.

La restricción nos dice que $x^{2}+y^{2}+z^{2} \leq 1$ lo que significa $|x|, |y|, |z| \leq 1$.

Para maximizar $x,y,z$ podemos elegir que $x=1$, $y=1$, $z=1$, entonces el máximo de $xyz$ se alcanza en $(1,1,1)$ y la función es $f(1,1,1)=1 \cdot 1 \cdot 1=1$

**Aquí podemos verlo en 2D**

```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt

# Definimos la función a maximizar
def f(x):
    return (x[0] * x[1] * x[2])

# Restricción por la ecuación del círculo
def constraint(x):
    return (x[0]**2 + x[1]**2 + x[2]**2 - 1)

# Análisis con curvas de nivel
x = np.linspace(-1.1, 1.1, 150)
y = np.linspace(-1.1, 1.1, 150)
X, Y = np.meshgrid(x, y)
Z = f(np.vstack([X.ravel(), Y.ravel(), Y.ravel()])).reshape((150, 150))

plt.figure(figsize=(8, 8))
plt.contour(X, Y, Z, np.arange(-5, 10, 0.05), cmap='jet')

# Restricción visualizada en la proyección XY
plt.plot(x, np.sqrt(1.0 - x**2), 'k:', linewidth=1)
plt.plot(x, -np.sqrt(1.0 - x**2), 'k:', linewidth=1)

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title("Proyección en el plano XY")
plt.axis([-1.1, 1.1, -1.1, 1.1])
```

**c)** maximizar $1-x^{2}-y^{2}$ sujeta a $x \ge 2$ y $y \ge 3$

Realizamos un bosquejo notamos que la función objetivo $1-x^{2}-y^{2}$ es una parábola abre hacia abajo. Dado que las restricciones $x \ge 2$ y $y \ge 3$ restringe el dominio a los valores de $x, y$ tenemos que $x \ge 2$ y $y \ge 3$.

Entonces queremos encontrar el punto en el cuarto cuadrante dónde cumple la restricción dads $x \ge 2$ y $y \ge 3$. En la restricción $x \ge 2$ nos lleva a qué $x=2$ (el valor mínimo permitido) y la restricción $y \ge 3$ nos lleva a qué $y=3$ (el valor mínimo permitido).

Entonces el punto que maximiza $1-x^{2}-y^{2}$ bajo las restricciones es:
$(2,3)$ y el valor máximo es:

$f(2,3)=1-2^{2}-3^{2}=1-4-9=12$.

```python
%matplotlib inline
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt

# Definimos la función a maximizar
def f(x):
    return (1 - x[0]**2 - x[1]**2)

# Restricciones de desigualdad
def constraint1(x):
    return x[0] - 2

def constraint2(x):
    return x[1] - 3

# Resolvemos la optimización con restricciones
result = opt.minimize(f, [2, 3], constraints=[{'type': 'ineq', 'fun': constraint1}, {'type': 'ineq', 'fun': constraint2}])

# Creamos una malla de puntos
x = np.linspace(0, 4, 50)
y = np.linspace(0, 4, 50)
X, Y = np.meshgrid(x, y)

# Evaluamos la función en la malla
Z = f(np.vstack([X.ravel(), Y.ravel()])).reshape((50, 50))

# Graficamos las curvas de nivel
plt.contour(X, Y, Z, np.arange(-20, 20, 0.5), cmap='jet')

# Graficamos las restricciones
plt.axvline(x=2, color='k', linestyle='dashed', label=r'$x => 2$')
plt.axhline(y=3, color='k', linestyle='dashed', label=r'$y => 3$')

# Marcamos el máximo encontrado
plt.plot(result.x[0], result.x[1], 'ro', label='Máximo encontrado')

# Establecemos los límites del gráfico
plt.axis([0, 3.5, 0, 3.5])

# Añadimos etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Mostramos el gráfico
plt.show()
```

**d)** Min $4In(x^{2}+2)+y^{2}$ sujeta a $x^{2}+y \ge 2$ y $x \ge 1$

Notamos que la función objetivo $4In(x^{2}+2)+y^{2}$ es una combinación de una función logarítmica y una función cuadrática. La restricción $x^{2}+y \ge 2$ nos dice que los puntos permitidos están fuera del círculo de radio $\sqrt 2$ centrados en el origen y la restricción $x \ge 1$ significa que los valores de $x$ son mayores o igual a 1.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función a maximizar
def f(x):
    return (4 * np.log(x[0]**2 + 2) + x[1]**2)

# Restricciones de desigualdad
def constraint1(x):
    return (x[0]**2 + x[1] - 2)

def constraint2(x):
    return (x[0] - 1)

# Creamos una malla de puntos
x = np.linspace(1, 3, 50)
y = np.linspace(0, 3, 50)
X, Y = np.meshgrid(x, y)

# Evaluamos la función en la malla
Z = f(np.vstack([X.ravel(), Y.ravel()])).reshape((50, 50))

# Graficamos las curvas de nivel
plt.contour(X, Y, Z, np.arange(-10, 20, 0.4), cmap='jet')

# Graficamos las restricciones
plt.axvline(x=1, color='k', linestyle='dashed', label=r'$x => 1$')
plt.plot([0.5, 0.5, 1.5, 1.5], [2.5, 1.5, 1.5, 2.5], 'k:', linewidth=1, label=r'$x^{2}+y => 2$')

# Rellenamos la región factible
plt.fill_betweenx(np.linspace(0, 3, 100), 1, 3, color='gray', alpha=0.3)

# Establecemos los límites del gráfico
plt.axis([0, 2.5, 0, 3.0])

# Añadimos etiquetas y leyenda
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Mostramos el gráfico
plt.show()
```
