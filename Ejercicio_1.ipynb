Un compañía fabrica dos tipos de motores pequeños con combustible sólido. Con el motor $A$ gana $\$3$ por motor y con el motor $B$ gana $\$4$ por motor.
Dispone de $80$ horas por semana para producirlos y en procesos se requieren: 4 horas para A y solo $2$ horas para $B$.
Sin embargo, debido a lo peligroso del material usado, son necesarios $5$ horas de preparación y limpieza para el motor $B$ y $2$ horas pra $A$.
El tiempo de preparación total disponible es $120$ horas por semana. Determinar el número de cada clase de motores a producir que de como resultado la maxima ganacia.

**De manera formal:**

Variables de decisión:

motor A, motor B, $x_1$, y $x_2$

**Función objetivo:**

maximizar: $z=3x_1+4x_2$

**restricciones:**

$x_1=$ Número de motores tipo A

$x_2=$ Número de motores tipo B

**Consideramos:**

$4x_1+2x_2\leq 80$

$2x_1+5x_2\leq 120$

$x_1\geq 0$

$x_2\geq 0$

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
#aquí vamos a dibujar la resticción
# Restricción para M1
plt.plot([0, 20.0], [40.0,0], color='pink', alpha=0.25)
plt.fill_between([0, 20.0], [40.0, 0], color='pink', alpha=0.25, label='resultante de M1')

plt.xlabel('recipientes K')
plt.ylabel('recipientes L')
plt.legend()
plt.show()
```
