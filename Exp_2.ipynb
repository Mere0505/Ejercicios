```python
#Este programa crea un gráfico 3D para representar la presión
#de un fluido y la tensión vertical en función de la profundidad de dos
#superficies distintas.
#primero: las bibliotecas que debemos usar:
#NumPY es para realizar calculos
#Matplotlib  para realizar gráficos
#mpl_toolkits.mplot3d para realizar los graficos en 3D

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definición de constantes
rho_w = 1000  # Densidad del agua en kg/m^3
g = 9.81  # Aceleración debida a la gravedad en m/s^2
lambda_value = 0.8  # Constante lambda

# Definición del rango de profundidades
depths = np.linspace(0, 200, 201)  # Desde la superficie hasta 200 metros de profundidad

# Calcula la presión y la tensión vertical en 3D
#aqui utilizamos Numpy para hacer la malla 2D (X, Y)
#para que presenten nuestras coordenadas de profundidad en el eje x, y.

X, Y = np.meshgrid(depths, depths)
pressures = rho_w * g * X
stresses = (1 - lambda_value) * pressures

# Crea una figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Grafica la presión y la tensión vertical en 3D
#Usamos el método plot_surface del subgráfico 3D (ax.plot_surface) para graficar
#las superficies 3D de presión y tensión vertical en función de las coordenadas de profundidad.

ax.plot_surface(X, Y, pressures, cmap='cool', label='Presión del fluido')
ax.plot_surface(X, Y, stresses, cmap='hot', label='Tensión vertical')

# Etiquetas de los ejes
ax.set_xlabel('Profundidad (m)')
ax.set_ylabel('Profundidad (m)')
ax.set_zlabel('Presión/Tensión')

# Mostrar el gráfico
plt.show()
```
