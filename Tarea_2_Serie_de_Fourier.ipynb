**Tarea 2:**
Calcular la serie de Fourier y graficar en una notebook de jupyter.

Para resolver los ejercicios, necesitamos considerar lo siguiente:

**1.** Se usará la formula símetrica de la serie de Fourier.

$ f(x) \thicksim \frac{1}{2}a_0 + \sum_{n=1}^{\infty}f(x)  \left[a_ncos\frac{n \pi x}{L}+b_n sen\frac{n \pi x}{L} \right]$

**2.**Para calcular series necesitamos las siguientes ecuaciones vistas en clase.

$a_0=\frac{1}{L}\int_{-L}^{L}f(x)dx$

$a_n=\frac{1}{L}\int_{-L}^{L}f(x) cos\frac{n \pi x}{L} dx, n=1, 2,\cdots$

$b_n=\frac{1}{L}\int_{-L}^{L}f(x) sen\frac{n \pi x}{L} dx, n=1, 2,\cdots$

     f(x) = \left\{

		 0 & 0 \le x < 1

```python
from pylab import *

def f(x):
    result = zeros(len(x))
    for i in range(len(x)):
        if -1 <= x[i] < 0:
            result[i] = 1
        elif 0 <= x[i] < 1:
            result[i] = 0
    return result

x = linspace(-1, 1, 1000)
f_values = f(x)

plot(x, f_values, label='$f(x)$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title("Ejercicio 1")
grid()
legend()
show()
```

```python
from pylab import *
import numpy as np

def f(x):
    y = np.zeros(len(x))
    for i in range(len(x)):
        if -1 <= x[i] < 0:
            y[i] = 1
        elif 0 <= x[i] < 1:
            y[i] = 0
    return y

def recta(x):
    y = 2*x + 1
    return y

x = linspace(-1, 1, 1000)
f_values = f(x)
plot(x, f_values, label='f(x)')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title("Función definida por partes")
xlim([-1.1, 1.1])
ylim([-0.1, 1.1])
grid()
legend()
show()

def u_0(x, N=20):
    f = np.zeros(len(x))
    for n in range(1, N):
        f += (1 - (-1)**n * 3) * sin(n * pi * x) / n
    return (2.0 / pi) * f

f_n = u_0(x)
plot(x, f_n, label='$f_N(x)$ con $N=20$')
plot(x, recta(x), label='Recta $2x+1$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title("Aproximación de la condición inicial")
xlim([-1.1, 1.1])
ylim([-0.1, 3.1])
grid()
legend()
show()

def u_x_t(x, t, N=50):
    f = np.zeros((len(x), len(t)))
    for i in range(len(t)):
        for n in range(1, N):
            f[:, i] += (1 - (-1)**n * 3) * sin(n * pi * x) / n * exp(-(n * pi)**2 * t[i])
    return (2.0 / pi) * f

x = linspace(-1, 1, 1000)
t = linspace(0, 10, 1000)
U = u_x_t(x, t)
plot(x, U[:, 0], label='$u(x,0)$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title("Aproximación de la condición inicial")
xlim([-1.1, 1.1])
ylim([-0.1, 3.1])
grid()
legend()
show()

# Representación gráfica en 3D
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection="3d")

x = linspace(-1.0, 1.0, 100)
t = linspace(0, 1.0, 100)

X, T = np.meshgrid(x, t)

U = u_x_t(x, t)

plt.xlabel('$t$')
plt.ylabel('$x$')
plt.title("u(x,t)")
ax.plot_surface(X, T, U, cmap="viridis")
plt.show()
```
